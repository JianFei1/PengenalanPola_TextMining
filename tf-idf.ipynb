## import data

```python
import pandas as pd 
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

df = pd.read_csv("hasil_textProcessing.csv")
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

## isi_berita_final to list

```python
list_isi_berita = []
for i in range(len(df.iloc[:, -1])):
    list_isi_berita.append(df.iloc[i, -1])
# print(list_isi_berita)
```

## tf-idf menggunakan TfidfTransformer

```python
count_vectorizer = CountVectorizer(min_df=1)
count_vectorizer.fit_transform(list_isi_berita)
```

```python
#get fitur
fitur = count_vectorizer.get_feature_names_out()
```

## term frequency

```python
# term frequency dengan tfidftransformer
tf = count_vectorizer.fit_transform(list_isi_berita).toarray()
df_tf =pd.DataFrame(data=tf,index=list(range(1, len(tf[:,1])+1, )),columns=[fitur])
df_tf.head(8)
```

```python
df_tf.to_csv('tf.csv')
```

```python
#tfidf dengan tfidf transformer
tfidf_transform = TfidfTransformer(use_idf=True,norm='l2',smooth_idf=True)
tfidf=tfidf_transform.fit_transform(count_vectorizer.fit_transform(list_isi_berita)).toarray()
df_tfidf =pd.DataFrame(data=tfidf,index=list(range(1, len(tfidf[:,1])+1, )),columns=[fitur])
df_tfidf.head(8)
```

```python
# df_tfidf.to_csv('tf-idf.csv')
```

```python
df_tfidf =  df_tfidf.reset_index()
df_tfidf.insert(0, 'judul_berita', df['judul'] )
df_tfidf.insert(1, 'label', df['label'] )
df_tfidf = df_tfidf.drop("index", axis=1)
```
