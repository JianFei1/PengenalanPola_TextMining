## TF-IDF

### import

```python
import pandas as pd 
import numpy as np

TWEET_DATA = pd.read_csv("hasil_textProcessing.csv")
TWEET_DATA = TWEET_DATA.drop("Unnamed: 0", axis=1)
TWEET_DATA = TWEET_DATA.drop("judul", axis=1)
TWEET_DATA = TWEET_DATA.rename(columns={"isi_berita_final": "tweet"})
TWEET_DATA.head()
```

```python
len_df = len(TWEET_DATA.index)

list_data_tweet = []
for h in range(len_df):
    data_index = TWEET_DATA.iloc[h, [1]].values
    listKata=[]
    kata=""
    for i in data_index[0]:
        if i != ' ':
            kata += i
        elif i == " ":
            listKata.append(kata)
            kata = ''
    list_data_tweet.append(listKata)
    
print(list_data_tweet)
```

```python
TWEET_DATA = TWEET_DATA.drop('tweet', axis=1)
TWEET_DATA.head()
```

```python
TWEET_DATA['tweet'] = np.array(list_data_tweet)
TWEET_DATA.head()
```

## to list

```python
# convert list formated string to list
import ast

def convert_text_list(texts):
    texts = ast.literal_eval(texts)
    return [text for text in texts]

TWEET_DATA["tweet_list"] = TWEET_DATA["tweet"].apply(convert_text_list)


# print(TWEET_DATA["tweet_list"][90])

# print("\ntype : ", type(TWEET_DATA["tweet_list"][90]))
```
